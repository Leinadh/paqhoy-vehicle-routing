```python
import os
import matplotlib.pyplot as plt
```

```python
dir_path_data_historica = 'data.historica.20210511'
ls_archivos_pedidos = []
for file_name in os.listdir(dir_path_data_historica):
#     print(file_name.split('.')[0])
    with open(os.path.join(dir_path_data_historica, file_name)) as f:
        lines = f.readlines()
        pedidos = [x[:-1] for x in lines]
        cant_pedidos = [int(x.split(',')[3]) for x in lines]
        
        # file_name, lista_pedidos, len_pedidos, max_paq_pedido, sum_paq
        ls_archivos_pedidos.append((file_name.split('.')[0], pedidos, len(pedidos), max(cant_pedidos), sum(cant_pedidos)))
        print(max(cant_pedidos))
#         print(len(lines))
ls_archivos_pedidos[0]
```

```python
ls_nombre_arch_pedidos = [x[0] for x in ls_archivos_pedidos]
ls_len_arch_pedidos = [x[2] for x in ls_archivos_pedidos]
ls_max_cant_paq = [x[3] for x in ls_archivos_pedidos]
ls_sum_cant_paq = [x[4] for x in ls_archivos_pedidos]
```

```python
print(len(ls_nombre_arch_pedidos))
print(ls_nombre_arch_pedidos)
```

```python
plt.plot(ls_nombre_arch_pedidos, ls_len_arch_pedidos)
plt.plot(ls_nombre_arch_pedidos, ls_max_cant_paq)
plt.plot(ls_nombre_arch_pedidos, ls_sum_cant_paq)
plt.legend(['Cantidad registros','Max cantidad paquetes','Suma cantidad paquetes'])
fecha_ini = ls_nombre_arch_pedidos[0][6:]
fecha_fin = ls_nombre_arch_pedidos[-1][6:]
plt.title(f'Registros de pedidos por día ({fecha_ini} - {fecha_fin})')
```

```python
ini_dia = 10
fin_dia = 25
plt.plot(ls_nombre_arch_pedidos[ini_dia:fin_dia+1], ls_len_arch_pedidos[ini_dia:fin_dia+1])
plt.plot(ls_nombre_arch_pedidos[ini_dia:fin_dia+1], ls_max_cant_paq[ini_dia:fin_dia+1])
plt.plot(ls_nombre_arch_pedidos[ini_dia:fin_dia+1], ls_sum_cant_paq[ini_dia:fin_dia+1])
plt.legend(['Cantidad registros','Max cantidad paquetes','Total cantidad paquetes'])
fecha_ini = ls_nombre_arch_pedidos[ini_dia][6:]
fecha_fin = ls_nombre_arch_pedidos[fin_dia][6:]
plt.title(f'Registros de pedidos por día ({fecha_ini} - {fecha_fin})')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'registro': ls_nombre_arch_pedidos, 'total_pedidos': ls_sum_cant_paq})
# df.to_csv('total_pedidos_por_dia.csv', index=False)
```

```python
df = df.reset_index()
```

```python
df.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
reg = linear_model.LinearRegression()
reg.fit(df['index'].values.reshape(-1, 1), df['total_pedidos'].values.reshape(-1, 1))
reg.coef_
```

```python
a = np.arange(100).reshape(-1,1)
array_pred_sum_cant_paq = reg.predict(a)
```

```python
ini_dia = 0
fin_dia = 99

# ls_pred_sum_cant_paq = [ls_sum_cant_paq[ini_dia]+m*(i_dia-ini_dia) for i_dia in range(ini_dia, fin_dia+1)]
plt.plot(ls_nombre_arch_pedidos[ini_dia:fin_dia+1], ls_sum_cant_paq[ini_dia:fin_dia+1])
# plt.plot(ls_nombre_arch_pedidos[ini_dia:fin_dia+1], ls_pred_sum_cant_paq)
plt.plot(ls_nombre_arch_pedidos[ini_dia:fin_dia+1], array_pred_sum_cant_paq[ini_dia:fin_dia+1])
plt.legend(['Total cantidad paquetes', "Predicción de crecimiento"])
fecha_ini = ls_nombre_arch_pedidos[ini_dia][6:]
fecha_fin = ls_nombre_arch_pedidos[fin_dia][6:]
plt.title(f'Registros de pedidos por día ({fecha_ini} - {fecha_fin})')
print('R2 score:',r2_score(df['total_pedidos'].values.reshape(-1,1), array_pred_sum_cant_paq))
```
